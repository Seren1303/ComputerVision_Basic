# Chapter 1.4.6 Image Gradients

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./img/mountain_sky.jpg',0)
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.figure(figsize=(16,12))
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./img/sudoku.jpg',0)
```

```python
# Output dtype = cv2.CV_8U
sobelx8u = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)
```

```python
# Output dtype = cv2.CV_64F. Then take its absolute and convert to cv2.CV_8U
sobelx64f = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
abs_sobel64f = np.absolute(sobelx64f)
sobel_8u = np.uint8(abs_sobel64f)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(sobelx8u,cmap = 'gray')
plt.title('Sobel CV_8U'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(sobel_8u,cmap = 'gray')
plt.title('Sobel abs(CV_64F)'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./img/lena.png',0)
plt.figure(figsize=(20,15))
edges = cv2.Canny(img,50,70)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('./img/mountain_sky.jpg')
img1 = cv2.pyrDown(img)
img2 = cv2.pyrDown(img1)
```

```python
cv2.imshow('image',img)
cv2.imshow('image1',img1)
cv2.imshow('image2',img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edge = cv2.Canny(img,100,200)
edge1 = cv2.Canny(img1,100,200)
edge2 = cv2.Canny(img2,100,200)
```

```python
cv2.imshow('image',edge)
cv2.imshow('image1',edge1)
cv2.imshow('image2',edge2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edge21 = cv2.pyrUp(edge1)
edge22 = cv2.pyrUp(edge2)
```

```python
cv2.imshow('edge21',edge21)
cv2.imshow('edge22',edge22)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
